```python
from collections import Counter
from itertools import product
import numpy as np
import pandas as pd
```

```python
aminoacids = pd.read_csv("./data/aminoacids_properties.csv")
aminoacids.set_index("Sym", inplace=True)
aminoacids
```

```python
aminoacids["Polarity"].value_counts()
```

```python
aminoacids["Polarity"].replace({"Nonpolar": 0, "Polar": 1, "Acidic polar": 1, "Basic polar": 1}, inplace=True)
```

```python
aminoacids["Charge"].value_counts()
```

```python
aminoacids["Charge"].replace({"Negative": 0, "Neutral": 0.5, "Positive": 1}, inplace=True)
```

```python
aminoacids = pd.concat([
    pd.get_dummies(aminoacids["Class"], prefix="Class"),
    aminoacids.iloc[:, 2:]
], axis=1)
aminoacids.drop("Class", axis=1, inplace=True)
aminoacids = (aminoacids - aminoacids.min()) / (aminoacids.max() - aminoacids.min())
aminoacids
```

```python
data_single_train = pd.read_csv("./data/single_muts_train.csv", index_col=0)
data_single_train
```

```python
data_single_test = pd.read_csv("./data/single_muts_test.csv", index_col=0)
data_single_test
```

```python
data_multi_train = pd.read_csv("./data/multiple_muts_train.csv", index_col=0)
data_multi_train
```

```python
to_drop = data_multi_train["stabilityscore"].index[data_multi_train["stabilityscore"].apply(np.isnan)]
data_multi_train.drop(to_drop, inplace=True)
```

```python
data_multi_test = pd.read_csv("./data/multiple_muts_test.csv", index_col=0)
data_multi_test
```

```python
def get_sequence_features(sequence: str):
    return np.vstack([
        aminoacids.T[s].values
        for s in sequence
    ])
```

```python
sequence_structure = pd.read_csv("./data/aminoacids_secondary_structure.csv")
sequence_structure.drop("Abbr.", axis=1, inplace=True)
sequence_structure.set_index("Sym", inplace=True)
sequence_structure = sequence_structure.div(sequence_structure.sum(axis=1), axis=0)
sequence_structure
```

```python
def get_seq_str_feat(sequence, structure):
    return np.array([
        sequence_structure.loc[a, s]
        for (a, s) in zip(sequence, structure)
    ])
```

```python
sequence_aminoacids = {
    aminoacid
    for sequence in data_single_train["sequence"]
    for aminoacid in sequence
}
```

```python
def extract_features(path, data: pd.DataFrame):
    # Extract sequence features
    sequence_feat = np.array([
        np.hstack([
            get_sequence_features(r["sequence"]),
            np.expand_dims(get_seq_str_feat(r["sequence"], r["secondary_structure"]), -1)
        ])
        for _, r in data.iterrows()
    ])
    # Export features
    np.savez_compressed(path, **{
        "sequence_feat": sequence_feat,
        "score": data["stabilityscore"]
    })
```

```python
extract_features("./data/data_single_train.npz", data_single_train)
```

```python
extract_features("./data/data_single_test.npz", data_single_test)
```

```python
extract_features("./data/data_multi_train.npz", data_multi_train)
```

```python
extract_features("./data/data_multi_test.npz", data_multi_test)
```
